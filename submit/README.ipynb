```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python

def get_section_results(file):
    
    X = []
    Y = []
    b_loss = []
    
    for i in tf.compat.v1.train.summary_iterator(file):
        for val in i.summary.value:
            if val.tag == 'Train_EnvstepsSoFar':
                X.append(val.simple_value)
            elif val.tag == 'Eval_AverageReturn':
                Y.append(val.simple_value)
            elif val.tag == 'Baseline_Loss':
                b_loss.append(val.simple_value)

            
    return X, Y, b_loss
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the moving average function
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Compute moving average for your data
window = 5  # Adjust this based on your preference

Y1_ma = moving_average(Y1, window)
Y2_ma = moving_average(Y2, window)
Y3_ma = moving_average(Y3, window)
Y4_ma = moving_average(Y4, window)
Y5_ma = moving_average(Y5, window)

b_loss1_ma = moving_average(b_loss1, window)
b_loss2_ma = moving_average(b_loss2, window)
b_loss3_ma = moving_average(b_loss3, window)
b_loss4_ma = moving_average(b_loss4, window)
b_loss5_ma = moving_average(b_loss5, window)

# Adjust the X values for moving average
X1_ma = X1[:-window+1]
X2_ma = X2[:-window+1]
X3_ma = X3[:-window+1]
X4_ma = X4[:-window+1]
X5_ma = X5[:-window+1]

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# First subplot for Baseline Loss
axes[0].plot(X3,Y3,label='lambda =0.99')
axes[0].plot(X3_ma,Y3_ma,label='lambda =0.99 moving average')
axes[0].legend()
axes[0].set_xlim([0, X1_ma[-1]])
axes[0].set_xlabel('Train_EnvStepsSoFar')
axes[0].set_ylabel('Baseline Loss (Moving Average)')
axes[0].set_title('Baseline Loss with Moving Average')

# Second subplot for Eval Average Return
axes[1].plot(X1_ma, Y1_ma, label='lambda=1')
axes[1].plot(X3_ma, Y3_ma, label='lambda=0.99')
axes[1].plot(X4_ma, Y4_ma, label='lambda=0.98')
axes[1].plot(X5_ma, Y5_ma, label='lambda=0.95')
axes[1].plot(X2_ma, Y2_ma, label='lambda=0')

axes[1].legend()
axes[1].set_xlim([0, X1_ma[-1]])
axes[1].set_xlabel('Train_EnvStepsSoFar')
axes[1].set_ylabel('Eval_AverageReturn (Moving Average)')
axes[1].set_title('Eval Average Return with Moving Average')

# Display the figure
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python

def get_section_results(file):
    
    X = []
    Y = []
    b_loss = []
    
    for i in tf.compat.v1.train.summary_iterator(file):
        for val in i.summary.value:
            if val.tag == 'Train_EnvstepsSoFar':
                X.append(val.simple_value)
            elif val.tag == 'Eval_AverageReturn':
                Y.append(val.simple_value)
            elif val.tag == 'Baseline_Loss':
                b_loss.append(val.simple_value)

            
    return X, Y, b_loss
```

```python
fname1 = 'data/q2_pg_cartpole_CartPole-v0_26-09-2023_09-46-43/events.out.tfevents.1695746803.MacBook-Pro-108.attlocal.net'
```

```python

X1, Y1, b_loss = get_section_results(fname1)
```

```python
plt.plot(X1,Y1)
```
